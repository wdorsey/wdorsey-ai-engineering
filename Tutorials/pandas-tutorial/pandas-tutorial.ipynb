## `DataFrame`
- A `DataFrame` is a 2-dimensional data structure that can store data of different types (including characters, integers, floating point values, categorical data and more) in columns. It is similar to a spreadsheet, a SQL table or the data.frame in R.

```python
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)
df
```

## `Series`
- Each column in a `DataFrame` is a `Series`
- When selecting a single column of a pandas `DataFrame`, the result is a pandas `Series`. To select the column, use the column label in between square brackets [].
- A pandas `Series` has no column labels, as it is just a single column of a `DataFrame`. A `Series` does have row labels.

```python
df["Age"]
```

```python
# manually create a series
ages = pd.Series([22, 35, 58], name="Age")
ages
```

```python
# get max Age from DataFrame
print(df["Age"].max())
# or from Series
print(ages.max())
```

## How do I select a subset of a `DataFrame`?

- To select a single column, use square brackets `[]` with the column name of the column of interest.

```python
ages = titanic["Age"]
ages.head()
```

- Each column in a `DataFrame` is a `Series`

```python
type(titanic["Age"])
```

- `DataFrame.shape` is an attribute of a pandas `Series` and `DataFrame` containing the number of rows and columns.

```python
print(titanic.shape)
print(titanic["Age"].shape)
```

- Selecting multiple columns returns a `DataFrame`.

```python
age_sex = titanic[["Age", "Sex"]]
print(type(age_sex))
print(age_sex.shape)
age_sex.head()
```

- Conditional expressions result in a boolean `Series`.

```python
titanic["Age"] > 35
```

- Filter the `DataFrame` by using the conditional `Series` in between the selection brackets `[]`.

```python
above_35 = titanic[titanic["Age"] > 35]
above_35
```

- `isin()` filters each row by checking if a given column value is in the given values.

```python
class_23 = titanic[titanic["Pclass"].isin([2, 3])]
class_23
```

- `notna()` filters only rows where the value is known.

```python
age_no_na = titanic[titanic["Age"].notna()]
age_no_na
```

- `loc` is used to both filter and select specific columns.

```python
adult_names = titanic.loc[titanic["Age"] > 35, "Name"]
adult_names
```

- `iloc` filters rows and columns by a given index range.

```python
titanic.iloc[9:25, 2:5]
```

- `loc`/`iloc` can also be used to set the value of rows/columns.

```python
# select the first 3 rows and 3rd column and set the value to "anonymous"
titanic.iloc[0:3, 3] = "anonymous"
titanic.head()
```

## How to create new columns derived from existing columns

- Calculations are done element-wise on all values in the given column.
- Both mathematical (`+`, `-`, `*`, `/`,...) and logical operators (`<`, `>`, `==`,...) work element-wise.
- `apply()` can be used to execute element-wise arbitrary python code.

```python
# London air quality in mg/m^3
air_quality["london_mg_per_cubic"] = air_quality["station_london"] * 1.882
air_quality.head()
```

```python
# Ratio of paris:antwerp
air_quality["ratio_paris_antwerp"] = (
    air_quality["station_paris"] / air_quality["station_antwerp"]
)
air_quality.head()
```

- Rename columns

```python
air_quality_renamed = air_quality.rename(
    columns={
        "station_antwerp": "BETR801",
        "station_paris": "FR04014",
        "station_london": "London Westminster",
    }
)
air_quality_renamed.head()
```

```python
# rename columns by running them through a function
air_quality_renamed = air_quality_renamed.rename(columns=str.lower)
air_quality_renamed.head()
```

### Long to wide table format

```python
# switch air_quality to air_quality_long.csv
air_quality = pd.read_csv(
    "data/air_quality_long.csv", index_col="date.utc", parse_dates=True
)
```

```python
air_quality.head()
```

```python
no2 = air_quality[air_quality["parameter"] == "no2"]
no2.head()
```

```python
no2_subset = no2.sort_index().groupby(["location"]).head(2)
no2_subset
```

### Pivot
- Use `pivot()` to turn each unique location into a column. row value is the "value" column.

```python
no2_subset.pivot(columns="location", values="value")
```

```python
no2.pivot(columns="location", values="value").plot()
```

- `pivot_table()` can be used to aggregate multiple values.
- The same 

```python
# index column is location
# columns are the unique parameter values
# row values are the mean of the value column
air_quality.pivot_table(
    values="value", index="location", columns="parameter", aggfunc="mean"
)
```

- Use `margins=True` to show subtotals for each row and column.

```python
air_quality.pivot_table(
    values="value",
    index="location",
    columns="parameter",
    aggfunc="mean",
    margins=True,
)
```

- The same result can be derived by using `groupby()`, but the resulting table is less readable.

```python
air_quality.groupby(["parameter", "location"])[["value"]].mean()
```

## Wide to long format

### `reset_index()`

```python
# location is the pivot column, so it is the new index.
# reset_index() overwrites the existing location values with a new sequential index
# the old date index is still included in the table as a column
no2_pivoted = no2.pivot(columns="location", values="value").reset_index()
no2_pivoted.head()
```

### `melt()`

```python
# turns all columns not included in id_vars into two columns:
#  - a column with the previous column header names
#  - a column with the values from those previous columns 
no_2 = no2_pivoted.melt(id_vars="date.utc")
no_2.head()
```

```python
# using explicit paramters
no_2 = no2_pivoted.melt(
    id_vars="date.utc",
    # which columns will be melted together
    value_vars=["BETR801", "FR04014", "London Westminster"],
    # name of value column
    value_name="NO_2",
    # name of melted column
    var_name="id_location",
)
no_2.head()
```

## How to combine data from multiple tables

```python
air_quality_no2 = pd.read_csv("data/air_quality_no2_long.csv", parse_dates=True)
air_quality_no2 = air_quality_no2[["date.utc", "location", "parameter", "value"]]
air_quality_no2.head()
```

```python
air_quality_pm25 = pd.read_csv("data/air_quality_pm25_long.csv", parse_dates=True)
air_quality_pm25 = air_quality_pm25[["date.utc", "location", "parameter", "value"]]
air_quality_pm25.head()
```

- Combine two tables with a similar structure with `concat()`
    - axis=0 combines rows

```python
air_quality = pd.concat([air_quality_pm25, air_quality_no2], axis=0)
air_quality.head()
```

```python
print('Shape of the air_quality_pm25 table: ', air_quality_pm25.shape)
print('Shape of the air_quality_no2 table: ', air_quality_no2.shape)
print('Shape of the air_quality table: ', air_quality.shape)
```

```python
air_quality = air_quality.sort_values("date.utc")
air_quality.head()
```

- The `keys` parameter allows for creating a new hierarchical index

```python
air_quality = pd.concat([air_quality_pm25, air_quality_no2], keys=["PM25", "NO2"])
air_quality
```

```python
# dropping the hierarchical index created above
air_quality.reset_index(level=0, drop=True)
```

## Join tables using a common identifier

```python
stations_coord = pd.read_csv("data/air_quality_stations.csv")
stations_coord.head()
```

- `merge()` performs a database-like join operation.

```python
air_quality = pd.merge(air_quality, stations_coord, how="left", on="location")
air_quality.head()
```

```python
air_quality_parameters = pd.read_csv("data/air_quality_parameters.csv")
air_quality_parameters.head()
```

```python
# specify columns of each DataFrame to join on
air_quality = pd.merge(
    air_quality, 
    air_quality_parameters, 
    how="left", 
    left_on="parameter", 
    right_on="id"
).drop("id", axis=1)
air_quality.head()
```

## How to handle time series data

```python
air_quality = pd.read_csv("data/air_quality_no2_long.csv").rename(columns={"date.utc": "datetime"})
air_quality.head()
```

```python
# this could be done while reading the file: read_csv("...", parse_dates=["datetime"])
air_quality["datetime"] = pd.to_datetime(air_quality["datetime"])
air_quality["datetime"]
```

```python
air_quality["datetime"].min(), air_quality["datetime"].max()
```

```python
air_quality["datetime"].max() - air_quality["datetime"].min()
```

```python
air_quality["month"] = air_quality["datetime"].dt.month
air_quality.head()
```

### What is the average NO2 concentration for each day of the week for each of the measurement locations?

```python
air_quality.groupby([air_quality["datetime"].dt.weekday, "location"])["value"].mean()
```

### Plot the typical NO2 pattern during the day of our time series of all stations together. In other words, what is the average value for each hour of the day?

```python
fig, axs = plt.subplots(figsize=(12, 4))
air_quality.groupby(air_quality["datetime"].dt.hour)["value"].mean().plot(
    kind="bar", rot=0, ax=axs
)
plt.xlabel("Hour of the day")
plt.ylabel("$NO_2 (µg/m^3)$")
```

### Datetime as index

```python
no_2 = air_quality.pivot(index="datetime", columns="location", values="value")
no_2.head()
```

```python
# datetime properties are now directly on the index
no_2.index.year, no_2.index.weekday
```

#### Create a plot of the NO2 values in the different stations from the 20th of May till the end of 21st of May

```python
no_2["2019-05-20":"2019-05-21"].plot()
```

### Resample a time series to another frequency

#### Aggregate the current hourly time series values to the monthly maximum value in each of the stations.
- `resample()` time series data to a different frequency.
  - Similar to groupby, it provies a time-based grouping via a string (`"ME"`).
  - Requires an aggregation function.

```python
monthly_max = no_2.resample("ME").max()
monthly_max
```

```python
# frequency is part of the index
monthly_max.index.freq
```

#### Make a plot of the daily mean NO2 value in each of the stations.

```python
no_2.resample("D").mean().plot(style="-o", figsize=(10, 5))
```
